```python
import subprocess
from pathlib import Path
import pandas as pd
import numpy as np

```

```python
batch_directory = Path("/mnt/storage/data/kas_k44a/ribosomes_21tomos/").absolute()

model_checkpoint_file = Path("/home/michalakdj/working/fidder_training/training/output/lightning_logs/version_0/checkpoints/epoch=33-step=600.ckpt").absolute()

pixel_size = 1.0825
probability_threshold = 0.5
```

```python
for ts_dir in batch_directory.glob("ts*"):
    ts_name = ts_dir.name

    output_mask = ts_dir / f"{ts_name}_fidder_mask_probthr{probability_threshold}.mrc"
    command = [
        "fidder",
        "predict",
        "--input-image",
        f"{ts_dir} / {ts_name}.mrc",
        "--model-checkpoint-file",
        f"{model_checkpoint_file}",
        "--pixel-spacing",
        f"{pixel_size}",
        "--probability-threshold",
        f"{probability_threshold}",
        "--output-mask",
        f"{output_mask}",
    ]
    #with open(ts_dir / f"{ts_name}_fidder_predict.log", "w") as log_file:
    result = subprocess.run(command) #, stdout=log_file, stderr=log_file)
    
    output_image = ts_dir / f"{ts_name}_fidder_erased_probthr{probability_threshold}.mrc"
    command = [
        "fidder",
        "erase",
        "--input-image",
        f"{ts_dir} / {ts_name}.mrc",
        "--input-mask",
        f"{output_mask}",
        "--output-image",
        f"{output_image}",
    ]
    #with open(ts_dir / f"{ts_name}_fidder_erase.log", "w") as log_file:
    result = subprocess.run(command) #, stdout=log_file, stderr=log_file)
```
