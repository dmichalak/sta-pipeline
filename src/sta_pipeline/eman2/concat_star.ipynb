```python

def concat_star(
    starfile_directory: Path,
    output_star: Path,
) -> None:
    """Concatenate all star files in a directory into one star file."""

    starfile_directory = Path(starfile_directory).absolute()
    output_star = Path(output_star).absolute()

    starfile_list = []
    for star in sorted(starfile_directory.glob("*seg*.star")):
        print(f"Reading {star.name}...")
        starfile_list.append(starfile.read(star))
    print("Concatenating star files...")
    concatenated_starfile_df = pd.concat(starfile_list, axis=0, ignore_index=True)
    print(f"Saving concatenated star file to {output_star}...")
    starfile.write(concatenated_starfile_df, output_star, overwrite=True)
    print("Done!")

    return concatenated_starfile_df
```

```python
starfile_directory = "/mnt/scratch/ribosomes/wws_EGFcontrol/eman2/segmentations/"
output_star = "/mnt/scratch/ribosomes/wws_EGFcontrol/eman2/segmentations/allpeaks_abs0.1rel0.1.star"

concat_star(
    starfile_directory,
    output_star,
)
```
