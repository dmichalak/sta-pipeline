```python
from pathlib import Path
import numpy as np
import pandas as pd

import starfile
```

```python
particle_directory = Path("/mnt/scratch/ribosomes/kas_k44a/dynamo_full/particles")

rln_angles = ["rlnAngleRot", "rlnAngleTilt", "rlnAnglePsi"]

for file in sorted(particle_directory.glob('*.star')):
    ts_name = file.name.split('_')[1] + '_' + file.name.split('_')[2]

    # Read star file
    df = starfile.read(file)
    
    # 
    df["rlnTomoName"] = ts_name

    # Randomize particle orientations
    for angle in rln_angles:
        random_angles = np.random.randint(0, 179, size=len(df.index))
        df[angle] = random_angles

    file_name = f"{file.stem}_fixed.star"
    starfile.write(df, particle_directory / file_name, overwrite=True)
```
